#### **Get Past _N_ Days of Player Data For All Players on Teams Playing Today**

```python
N = 14
PlayerStatsbyTeam = NDaysRecencyPlayerData(Game_DF, N)
```

```python
team = 'Chicago Cubs'
```

#### **Homerun Analysis over past _N_ Days**

```python
HR_over = 2


if team == 'ALL':
    PlayerStats = PlayerStatsbyTeam
else:
    PlayerStats = PlayerStatsbyTeam[team]

for player, stats in PlayerStats.items():
    
    if "HR" not in player:
        continue
    if sum(stats) >= HR_over:
        name = player.split(' - ')[0]
        print('{name} has hit {HRs} HRs in the past {N} days'.format(name=name, HRs=sum(stats), N=N))
        print(stats, '\n')
        
```

#### **Batting Average Analysis over past _N_ Days**

```python
overBA = 0.300
increasingDays = 3

if team == 'ALL':
    PlayerStats = PlayerStatsbyTeam
else:
    PlayerStats = PlayerStatsbyTeam[team]

for player, stats in PlayerStats.items():
    player_FLAG = 0

    if "AVG" not in player:
        continue
    
    name = player.split(' - ')[0]
    stats = [float(i) for i in stats]
    if mean(stats) >= overBA:
        player_FLAG = 1
        print('{name} has a {AVG} AVG for the past {N} days'.format(name=name, AVG=round(mean(stats),3), N=N))

    prev = stats[0]
    streak = 0
    maxstreak = 0
    for i in range(1,len(stats)):
        curr = stats[i]
        if curr > prev and curr != 0:
            streak += 1
        elif curr == prev and curr !=0:
            continue
        else:
            streak = 0

        maxstreak = max(maxstreak, streak)
        prev = curr

    if maxstreak >= increasingDays:
        player_FLAG = 1
        print('{name} had a {maxstreak} day streak of increasing AVG in the past {N} days'.format(name=name, maxstreak=maxstreak, N=N))
    
    if player_FLAG:
        print(stats, '\n')

        
```
